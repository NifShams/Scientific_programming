```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre, jv

# Plot 1: Legendre Polynomial of Degree 6
x_0 = np.linspace(0, 1, 100)
plt.plot(x_0, legendre(6)(x_0), label="Legendre Polynomial (n=6)", color="blue")
plt.title("Legendre Polynomial (Degree 6)")
plt.xlabel("x")
plt.ylabel("P₆(x)")
plt.legend()
plt.grid()
plt.show()

# Plot 2: Bessel Function of the First Kind (Order 3)
x_1 = np.linspace(0, 10, 100)
plt.plot(x_1, jv(3, x_1), label="Bessel Function J₃(x)", color="green")
plt.title("Bessel Function of the First Kind (Order 3)")
plt.xlabel("x")
plt.ylabel("J₃(x)")
plt.legend()
plt.grid()
plt.show()

```
