```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.set_theme(style="darkgrid")

# Bar Plot
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", errorbar="sd")
plt.title("Average Total Bill by Day")
plt.show()

# Count Plot
sns.countplot(data=tips, x="day", hue="day", palette="coolwarm", legend=False)
plt.title("Count of Days")
plt.show()

# Box Plot
sns.boxplot(data=tips, x="day", y="total_bill", hue="day", palette="Set2", legend=False)
plt.title("Box Plot of Total Bills")
plt.show()

# Violin Plot
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True, palette="muted")
plt.title("Violin Plot of Total Bills")
plt.show()


penguins = sns.load_dataset("penguins")
sns.set_theme(style="darkgrid")

sns.barplot(data=penguins, x="species")
plt.title("Penguins")
plt.show()


sns.barplot(x="sex", y="bill_length_mm", data=penguins )
plt.title("Average Bill Length for Each Sex in Penguins Dataset")
plt.show()



plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  
sns.boxplot(x="time", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill Grouped by Time")


plt.subplot(1, 2, 2) 
sns.violinplot(x="time", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill Grouped by Time")

# Display the plots
plt.tight_layout()
plt.show()
```
