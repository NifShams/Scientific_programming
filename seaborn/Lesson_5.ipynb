```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
tips = sns.load_dataset("tips")
penguins = sns.load_dataset("penguins")
sns.set(style="darkgrid")

# Select numeric columns from both datasets
numeric_tips_0 = tips.select_dtypes(include=['float64', 'int64'])
numeric_tips_1 = penguins.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrices
corr_0 = numeric_tips_0.corr()
corr_1 = numeric_tips_1.corr()

# Set up the figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

# Plot the first heatmap (tips dataset)
sns.heatmap(corr_0, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, ax=axes[0])
axes[0].set_title("Correlation Heatmap - Tips Dataset")

# Plot the second heatmap (penguins dataset)
sns.heatmap(corr_1, annot=True, cmap="viridis", fmt=".3f", linewidths=0.5, ax=axes[1])
axes[1].set_title("Correlation Heatmap - Penguins Dataset")

# Show the plot
plt.tight_layout()  
plt.show()
```


### 1. `corr = tips.corr()`
- This line computes the **correlation matrix** of the `tips` dataset, which contains numeric data like `"total_bill"`, `"tip"`, and `"size"`.
- The `.corr()` method calculates the **Pearson correlation coefficient** between pairs of numeric columns. The result is a square matrix where each cell represents the correlation between two variables:
  - A value of `1` means a perfect positive correlation (as one variable increases, the other increases).
  - A value of `-1` means a perfect negative correlation (as one variable increases, the other decreases).
  - A value of `0` means no correlation.
  
### 2. `sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)`
- **`sns.heatmap()`** is used to generate a heatmap based on the correlation matrix (`corr`).
- **`annot=True`**: Annotates each cell of the heatmap with the correlation value, so you can see the numeric value of the correlation directly in the heatmap.
- **`cmap="coolwarm"`**: Defines the color map for the heatmap. The "coolwarm" palette uses blue for negative correlations and red for positive correlations. The colors gradually shift from blue (low values) to red (high values).
- **`fmt=".2f"`**: Formats the numeric values in each cell to 2 decimal places.
- **`linewidths=0.5`**: Adds lines between the cells in the heatmap with a width of 0.5 to make the cells visually distinct from one another.

### 3. `plt.title("Correlation Heatmap")`
- This sets the **title** of the heatmap to `"Correlation Heatmap"`, which will be displayed at the top of the plot.

### 4. `plt.show()`
- This displays the heatmap on the screen.

### Result:
- You will see a heatmap where the numeric values in the `tips` dataset are correlated with each other. Each cell in the heatmap will have a color indicating the strength and direction of the correlation between two variables, with the correlation values annotated inside each cell.
- The title will show at the top of the plot to indicate what the heatmap represents.
