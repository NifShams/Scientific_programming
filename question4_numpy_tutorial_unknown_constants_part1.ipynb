```python
import numpy as np
import matplotlib.pyplot as plt

# Define dimensionless time range
tilde_t = np.linspace(0, 10, 1000)  # Dimensionless time

# Dimensionless temperature function
tilde_T = 1 / (1 + np.exp(-tilde_t))

# Dimensionless power
tilde_P = tilde_T**4

# Compute dimensionless energy
tilde_E = np.cumsum(tilde_P * np.diff(np.append(tilde_t, tilde_t[-1])))  # Numerical integration

# Plot results
plt.figure(figsize=(10, 5))

# Plot dimensionless power
plt.subplot(1, 2, 1)
plt.plot(tilde_t, tilde_P, label="Dimensionless Power (P~)")
plt.xlabel("Dimensionless Time (t~)")
plt.ylabel("Dimensionless Power")
plt.title("Dimensionless Power vs. Time")
plt.grid(True)
plt.legend()

# Plot dimensionless energy
plt.subplot(1, 2, 2)
plt.plot(tilde_t, tilde_E, label="Dimensionless Energy (E~)", color="orange")
plt.xlabel("Dimensionless Time (t~)")
plt.ylabel("Dimensionless Energy")
plt.title("Dimensionless Energy vs. Time")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
