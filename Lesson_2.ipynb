```python

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.set_theme(style="darkgrid")

# Histogram
sns.histplot(tips["total_bill"], bins=15, kde=False, color="green")
plt.title("Histogram")
plt.show()

# KDE Plot
sns.kdeplot(tips["total_bill"], fill=True, color="red")
plt.title("KDE Plot")
plt.show()

# Combined Plot
sns.displot(tips, x="total_bill", kind="kde", hue="day", fill=True)
plt.title("KDE Plot by Day")
plt.show()




data = sns.load_dataset("iris")
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Width by Species")
plt.show()




sns.histplot(data["sepal_length"], bins=15, kde=False, color="green" , element="poly")
plt.title("Histogram Sepal Poly")
plt.show()



sns.histplot(data["sepal_length"], bins=15, kde=False, color="green" , element="step")
plt.title("Histogram Sepal Step")
plt.show()


sns.histplot(data["sepal_length"], bins=15, kde=False, color="green" )
plt.title("Histogram Sepal Default")
plt.show()
```
