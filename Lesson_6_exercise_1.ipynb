```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")
sns.set(style="darkgrid")

# Create a FacetGrid
g = sns.FacetGrid(penguins, col="species", height=5, aspect=1.2)

# Map boxplots to the FacetGrid
g.map(sns.boxplot, "sex", "bill_length_mm", order=["Male", "Female"])

# Set axis labels and titles
g.set_axis_labels("Sex", "Bill Length (mm)")
g.set_titles("{col_name} Species")

# Show the plot
plt.show()
```
