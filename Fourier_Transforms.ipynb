```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generate the time axis
time = np.linspace(0, 10 * np.pi, 100)  # 100 points between 0 and 10*pi
sampling_interval = time[1] - time[0]  # Calculate the time between samples

# Create the signal: two sine waves and random noise
signal = np.sin(2 * np.pi * time) + np.sin(4 * np.pi * time) + 0.1 * np.random.randn(len(time))

# Plot the time-domain signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, signal)
plt.title('Time Domain Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

# Perform the Fourier Transform (FFT)
N = len(signal)
fft_result = fft(signal)[:N//2]  # Use only the positive half of the FFT output
frequencies = fftfreq(N, sampling_interval)[:N//2]  # Corresponding frequencies

# Plot the frequency-domain spectrum
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(fft_result))
plt.title('Frequency Domain Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')

# Display the plots
plt.tight_layout()
plt.show()
```
