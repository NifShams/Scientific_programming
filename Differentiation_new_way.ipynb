```python
import numpy as np
import matplotlib.pyplot as plt
import numdifftools as nd

# Define the function
def f(x):
    return x ** 2 * np.sin(2 * x) * np.exp(-x)

# Define the range for x
x = np.linspace(0, 1, 100)

# Compute the derivatives using numdifftools
df_dx_1 = nd.Derivative(f, n=1)(x)  # First-order derivative
df_dx_2 = nd.Derivative(f, n=2)(x)  # Second-order derivative
df_dx_3 = nd.Derivative(f, n=3)(x)  # Third-order derivative

# Plot the function and its derivatives
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df_dx_1, label="First-order derivative")
plt.plot(x, df_dx_2, label="Second-order derivative")
plt.plot(x, df_dx_3, label="Third-order derivative")

# Add legend and labels
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Function and its Derivatives")
plt.grid(True)
plt.show()
```
