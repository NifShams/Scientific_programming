```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")
sns.set(style="darkgrid")

# Create a FacetGrid
g = sns.FacetGrid(penguins, col="species", hue="sex", height=5, aspect=1.2)

# Map KDE plots to the FacetGrid
g.map(sns.kdeplot, "bill_length_mm", fill=True, alpha=0.5)


g.add_legend(title="Sex")


g.set_axis_labels("Bill Length (mm)", "Density")
g.set_titles("{col_name} Species")


plt.show()

```
