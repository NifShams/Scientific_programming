```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2,2,1000)
y = np.linspace(-2,2,1000)



xv , yv = np.meshgrid(x , y)
f = np.exp(-((xv**2) + (yv**2))) * np.sin(xv)
plt.contourf(xv , yv , f, levels=100)



abs_area = (np.abs(f.ravel()).sum()) * (x[1] -x [0] ) *( np.diff(y)[0])
print(abs_area)




new = np.sqrt((xv**2) + (yv**2))

abs_area_new = (np.abs(f.ravel()[ new.ravel() > 0.5]).sum()) * (x[1] -x [0] ) *( np.diff(y)[0])
print(abs_area_new)


```
