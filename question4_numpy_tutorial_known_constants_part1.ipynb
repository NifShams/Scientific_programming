```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
A = 1          # Surface area (arbitrary unit)
sigma = 5.67e-8 # Stefan-Boltzmann constant (W/m^2/K^4)
epsilon = 1    # Emissivity
T0 = 5000      # Initial temperature (Kelvin)
k = 0.1        # Rate constant (1/s)

# Define time range
t = np.linspace(0, 100, 1000)  # Time from 0 to 100 seconds

# Define temperature function
T = T0 / (1 + np.exp(-k * t))

# Define power function
P = A * sigma * epsilon * T**4

# Perform numerical integration to calculate total energy
E = np.cumsum(P * np.diff(np.append(t, t[-1])))  # Numerical integration using trapezoidal rule

# Plot results
plt.figure(figsize=(10, 5))

# Plot power
plt.subplot(1, 2, 1)
plt.plot(t, P, label="Power (P)")
plt.xlabel("Time (s)")
plt.ylabel("Power (W)")
plt.title("Power vs. Time")
plt.grid(True)
plt.legend()

# Plot total energy
plt.subplot(1, 2, 2)
plt.plot(t, E, label="Total Energy (E)", color="orange")
plt.xlabel("Time (s)")
plt.ylabel("Energy (J)")
plt.title("Total Energy vs. Time")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()



 
```
