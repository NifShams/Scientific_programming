```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

#Interpolation

from scipy.interpolate import interp1d


a = np.linspace(0, 10, 10)  # Original x-values
b = a**2 * np.sin(a)        # Original y-values


plt.scatter(a, b, label="Original Data", color="black")


i_0 = interp1d(a, b, kind='linear')
a_dense_0 = np.linspace(0, 10, 100)  # Dense x-values for interpolation
b_dense_0 = i_0(a_dense_0)
plt.plot(a_dense_0, b_dense_0, label="Linear Interpolation", linestyle="--")


i_1 = interp1d(a, b, kind='cubic')
b_dense_1 = i_1(a_dense_0)
plt.plot(a_dense_0, b_dense_1, label="Cubic Interpolation", linestyle="-.")


plt.xlabel("a")
plt.ylabel("b")
plt.title("Interpolation Comparison")
plt.legend()
plt.grid()


plt.show()

```
