```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (Seaborn dataset or your custom data)
data = sns.load_dataset("penguins")

# List of themes to apply
themes = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# Create the plot with each theme
for theme in themes:
    sns.set(style=theme)
    
    # Create the plot (e.g., a histogram)
    sns.histplot(data["bill_length_mm"], kde=True, bins=10, color="blue")
    
    # Add titles and labels
    plt.title(f"Distribution of Bill Length - Theme: {theme}")
    plt.xlabel("Bill Length (mm)")
    plt.ylabel("Frequency")
    
    # Display the plot
    plt.show()



```
