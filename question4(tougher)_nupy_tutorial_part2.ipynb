```python
import numpy as np

N = 1000
x = np.linspace( 0 ,1 , N+1)
dx = x[1] - x[0]
main_diag = -2 * np.ones(N-1)
off_diag = np.ones(N-2)

derivative_matrix = (np.diag(main_diag) + np.diag(off_diag , -1)  + np.diag(off_diag , 1)) / dx **2
x2_matrix = np.diag((10 * x[1:-1])**2)

LHS = derivative_matrix + x2_matrix

eig_value , eig_vector = np.linalg.eigh(LHS)
sorted_indices = np.argsort(eig_value)
eig_values = eig_value[sorted_indices]
eig_vectors = eig_vector[:, sorted_indices].T

print(eig_values)
```
