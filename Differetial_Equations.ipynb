```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
# odeint takes a derivative , initial codition, and the variable with respect to which the derivative is being taken


# Air friction while falling

def dv_dt(v, t):
    return 3 * v**2 - 5

v0 = 0

t = np.linspace(0 ,1 , 100)
sol_0 = odeint(dv_dt , v0 , t) 
sol_0 = sol_0.T
plt.plot(t , sol_0[0], label="Velocity vs Time (Air Friction)")
plt.xlabel('Time (t)')
plt.ylabel('Velocity (v)')
plt.title('Air Friction While Falling')
plt.legend()
plt.show()


# Coupled first-order ODEs

def dS_dx(S, x):
    y1, y2 = S
    return [y1 + y2**2 + 3*x, 3*y1 + y2**3 - np.cos(x)]

y1_0 = 0
y2_0 = 0
S_0 = (y1_0, y2_0)

x = np.linspace(0, 1, 100)
sol_1 = odeint(dS_dx, S_0, x)
sol_1 = sol_1.T
y1_sol = sol_1[0]
y2_sol = sol_1[1]

# Plot each solution separately
plt.plot(x, y1_sol, label="y1(x)", color='b')
plt.plot(x, y2_sol, label="y2(x)", color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution to Coupled First-Order ODEs')
plt.legend()
plt.show()



# Second Order ODEs

#Scipy can only solve coupled first order ODEs, but any second order ODE can be turned into two coupled first order ODEs. The same thing goes for higher order ODEs.


def dP_dT(P , T):
    w , theta = P
    return [np.sin(theta) , w ]

theta_0 = np.pi / 6
w_0 = 0 
P_0 = (w_0 , theta_0)

# Time array
T = np.linspace(0, 20, 200)

# Solve the system of ODEs
sol_2 = odeint(dP_dT, P_0, T)

# Separate the solution into w(t) and theta(t)
sol_2 = sol_2.T
w_sol = sol_2[0]
theta_sol = sol_2[1]

# Plot theta(t) as a function of t
plt.plot(T, theta_sol, label="theta(t)", color='b')
plt.xlabel('Time (t)')
plt.ylabel('Theta (θ)')
plt.title('Theta vs Time')
plt.legend()
plt.show()

# Plot w(t) as a function of t
plt.plot(T, w_sol, label="w(t)", color='r')
plt.xlabel('Time (t)')
plt.ylabel('Angular Velocity (w)')
plt.title('Angular Velocity vs Time')
plt.legend()
plt.show()
```
