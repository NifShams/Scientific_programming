```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.misc import derivative


def f(x):
    return x ** 2 * np.sin(2 * x) * np.exp( -x )
x = np.linspace(0,1,100)
df_dx_1 = derivative(f, x, dx =1e-6 , n=1) # first order derivative
df_dx_2 = derivative(f, x, dx =1e-6 , n=2) #second order derivative
#df_dx_2 = derivative(f, x, dx =1e-6 , n=3) #throws an error

plt.plot(x , f(x))
plt.plot(x , df_dx_1)
plt.plot(x , df_dx_2)
#plt.plot(x , df_dx_3)
```
